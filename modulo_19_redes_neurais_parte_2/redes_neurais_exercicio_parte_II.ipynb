```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split

import keras
from keras.layers import Input, Dense
from keras.models import Model, Sequential
from keras import metrics
from keras.callbacks import History 

import tensorflow as tf
```

```python
# Define exemplos de dados de treino
num_training = 1000
```

```python
# Define o tamanho da entrada
input_shape = 10
```

```python
# Gera inputs aleatórios para os dados de treino
x_train = np.random.rand(num_training, input_shape)
y_train = np.random.rand(num_training, 1)
```

```python
# Define a entrada da rede neural
inputs = tf.keras.Input(shape=(input_shape,))
```

```python
# Adiciona uma camada com 32 neurônios e uma função de ativação ReLU
x = tf.keras.layers.Dense(32, activation='relu')(inputs)
```

```python
# Adiciona uma segunda camada com 16 neurônios e uma função de ativação sigmóide
x = tf.keras.layers.Dense(16, activation='sigmoid')(x)
```

```python
# Adiciona uma camada de saída com um único neurônio e uma função de ativação linear
outputs = tf.keras.layers.Dense(1, activation='linear')(x)
```

```python
# Cria o modelo a partir das camadas de entrada e saída
model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
# Compila o modelo com a função de perda mean squared error e o otimizador Adam
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Treina o modelo usando os dados de treinamento
model.fit(x_train, y_train, epochs=100, batch_size=32)
```

```python
model.get_weights()
```

```python
# Previsões da rede
y_pred = model.predict(x_train)
```

```python
# Salvando o modelo
model.save("rede_train.h5")
```

**Atividade módulo 3 | Intro redes neurais parte 2**

Pegar uma rede simples e adicionar/alterar tipos de inicialização e funções de ativação, reutilização de camadas pré-treinadas, aceleração e otimização de hiperparametros.

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.initializers import glorot_uniform
from tensorflow.keras.models import Model

# Especificando a entrada
inputs = Input(shape=(1,))

# Adicionando uma camada com o parâmetro kernel_initializer para testar outra forma de inicialização. Além disso, poderíamos
# alterar a função de ativação desta camada, alterando o parâmetro "activation"
x = Dense(1, activation='linear', kernel_initializer=glorot_uniform())(inputs)

# Criando o modelo a partir das camadas
model = Model(inputs=inputs, outputs=x)
```

```python
# Criando nova entrada para o novo modelo com a mesma forma dos dados de entrada da camada que será reutilizada
new_inputs = tf.keras.Input(shape=(input_shape,))
```

```python
# Pegando a 1ª camada do modelo antigo
previous_layer = model.layers[0]
```

```python
# Congelando os pesos da camada para não alterá-los durante o treinamento
previous_layer.trainable = False
```

```python
# Conectando a nova entrada à camada do modelo antigo
x = previous_layer(new_inputs)
```

```python
# Adicionando uma nova camada após a camada reutilizada
x = tf.keras.layers.Dense(16, activation='sigmoid')(x)
```

```python
# Adicionando camada de saída
outputs = tf.keras.layers.Dense(1, activation='linear')(x)
```

```python
# Criando novo modelo a partir das camadas de entrada e saída
new_model = tf.keras.Model(inputs=new_inputs, outputs=outputs)
```

```python
# Compilando o modelo com a função de perda mean squared error e o otimizador Adam
new_model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Treinando o modelo com os dados de treino
new_model.fit(x_train, y_train, epochs=100, batch_size=32)
```

```python
# Previsões da rede
y_pred = new_model.predict(x_train)
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV

# Função para otimizar hiperparametros testando diferentes combinações de função de ativação, otimizador, 
# tamanho de batch e número de épocas. O objeto GridSearchCV recebe o modelo, o dicionário de parâmetros 
# que serão testados e o número de folds do cross-validation. O método fit() executa a busca e os melhores
# hiperparametros são armazenados no grid_result:

def create_model(activation='relu', optimizer='adam'):
    model = Sequential()
    model.add(Dense(32, input_shape=(input_shape,), activation=activation))
    model.add(Dense(16, activation='sigmoid'))
    model.add(Dense(1, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model

model = KerasRegressor(build_fn=create_model)

param_grid = {
    'activation': ['relu', 'sigmoid'],
    'optimizer': ['adam', 'rmsprop'],
    'batch_size': [16, 32, 64],
    'epochs': [50, 100, 200]
}

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(x_train, y_train)
```
