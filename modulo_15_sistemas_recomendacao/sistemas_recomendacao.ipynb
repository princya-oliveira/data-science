### Métodos baseados em conteúdo

<br>

Os métodos baseados em conteúdo usam __variáveis a respeito dos usuários e/ou dos itens__ que queremos recomendar. 

Por exemplo, considere que queremos fazer uma recomendação de filmes. Podemos usar:

    - informações dos usuários: sexo, idade, trabalho, etc 
    
    - informações dos filmes: tema principal (ação/comédia/..), ator principal, data de lançamento, etc
    
E dessa forma, utilizando essas informações, gerar recomendações direcionadas para públicos específicos.

Na prática, os __modelos criados são baseados nessas informações disponíveis__ e podem ser criados modelos tanto de regressão quanto classificação.

Uma __dificuldade__ em se utilizar esses métodos é justamente ter as informações!  

Não iremos entrar em detalhes a respeito dos métodos baseados em conteúdo neste módulo. De qualquer forma, links interessantes a respeito desses métodos serão disponibilizados no chat do curso!

### Filtros colaborativos

<br>

Filtros colaborativos são métodos baseados apenas __nas interações passadas dos usuários e dos itens__ que queremos recomendar. Desta forma, neste caso, __não é necessário termos informações dos usuários/itens!__ 

Aqui, apenas iremos precisar da __matriz de interação usuário/item:__ 

    - Considere uma empresa que possui N usuários em seu sistema. Além disso, considere que essa empresa vende vários produtos (itens), digamos que essa quantidade seja M. 
    
    - O objetivo de um sistema de recomendação é, para cada um dos N usuários, descobrir quais dos M produtos são os mais "interessantes" para cada usuário.
    
    - Analisando os dados históricos, podemos mensurar interações dos usuários com os itens. Alguns exemplos:
    
        - Se um usuário comprou ou não o item;
        
        - Qual a nota (de 1 a 10, por ex.) que o usuário deu para um item;
        
    - A partir dessas interações, podemos salvá-las numa matriz:
    
        - Cada linha dessa matriz representa um usuário;
        
        - Cada coluna dessa matriz representa um item;
        
        - As entradas da matriz registram as interações dos usuários com os itens.
        
        
<br>

Na prática, é natural que os usuários tenham interações com apenas parte dos itens. Dessa forma, a matriz de interação usuário/item acaba tendo muitas entradas com valores faltantes.

Essa propriedade é conhecida como __esparsidade__ da matriz de interação.

<br>

__Por exemplo,__ imagine uma empresa cujos __itens são filmes__ e as __interações são as notas que os usuários deram para os filmes que assistiram.__ Naturalmente, para cada usuário, muitos filmes não foram assistidos e portanto não terão uma nota associada, o que gera um valor faltante na matriz de interação.

<br>
<br>

<br><br>

A principal __motivação__ dos filtros colaborativos é que as __interações passadas entre usuários/itens__ é o suficiente para detectar __similaridade__ entre os mesmos e assim fazer as predições.

Neste contexto, iremos discutir bastante a respeito de como calcular similaridades, seja entre usuários ou similaridade entre itens.

<br>

__Tipos de filtros colaborativos:__

Os filtros __baseados em memória__ não usam modelos de aprendizado. Na prática, eles são __baseados nos cálculos de similaridade.__ 

São divididos entre:

     - filtros baseados nos usuários, onde a similaridade é calculada a nível usuário;
     
     - filtros baseados nos itens, onde a similaridade é calculada a nível item;
     
     
Por outro lado, os filtros __baseados em modelos__ terão aprendizado. Na prática, discutiremos __sistemas de recomendação baseados em fatoração de matrizes__ e usaremos o __gradient descent__ na etapa de aprendizado!

<br>

__As principais vantagens__ dos filtros colaborativos é que eles não dependem de informações dos usuários/itens, apenas fazem uso da matriz de interação entre os mesmos. Dessa forma, __podem ser usados em vários contextos diferentes!__

Além disso, a medida que os __usuários vão interagindo__ mais com os itens, o __sistema vai ficando cada vez mais efetivo.__

No entanto, existe um __trade-off:__ Problema do "início frio" (ou __cold start__, no inglês)

    - Os filtros colaborativos fazem uso apenas de interações passadas dos usuários (ou seja, apenas utilizam a matriz de interação usuário/item para fazer as recomendações) e isso gera um problema: é impossível recomendar itens para usuários novos (ou seja, usuários que ainda não tiveram nenhuma interação com os itens) ou recomendar um item novo para usuários (ou seja, um item que ainda não teve nenhuma interação com nenhum usuário).
    
    - Outro problema é que muitos usuários ou itens acabam tendo poucas interações, o que dificulta recomendações precisas.
    
Para contornarmos o problema do cold start, algumas iniciativas podem ser:

    - Estratégia aleatória: recomendar itens aleatórios para usuários novos ou recomendar um item novo para usuários aleatórios;
    
    - Estratégia da máxima expectativa: recomendar os melhores itens para usuários novos ou recomendar um item novo para os melhores usuários. 
    
    - Utilizar outro método de recomendação no "início da vida" do usuário e/ou do item.

Antes de falarmos sobre similaridade, vamos relembrar a __distância entre dois pontos:__

<br>

<br>

Na figura acima, estamos calculando a distância entre dois pontos $X = (X1, X2)$ e $Y = (Y1, Y2)$.

Podemos, naturalmente, generalizar e calcular a distância entre pontos em mais dimensões:

Considere $X = (X1, X2, .., Xn)$ e $Y = (Y1, Y2, .., Yn)$, onde n $\geq $2. Definimos:

$Dist(X, Y) = \sqrt{(X1 - Y1)^2 + (X2 - Y2)^2 + .. + (Xn - Yn)^2 }$.

<br>

Vamos implementar no python esse cálculo.

<br>

Na prática, usaremos a função do sklearn no link:

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.nan_euclidean_distances.html

```python
from sklearn.metrics.pairwise import nan_euclidean_distances

def euclid_dist(x, y):
    v = np.vstack([x, y])
    if np.isnan(nan_euclidean_distances(v)[0,1]):
        return 1e10
    else:
        return nan_euclidean_distances(v)[0,1]
```

```python
x = np.array([1,1])
y = np.array([2,2])

dist_xy = euclid_dist(x, y)

plt.title("Distância entre x e y: " + str(np.round(dist_xy, 3)), fontsize = 16, fontweight = 'bold')
plt.scatter([x[0], y[0]], [x[1], y[1]], s = 100, c = 'black')
plt.xlim(0,10)
plt.ylim(0,10)
plt.xticks(range(11))
plt.yticks(range(11))
plt.grid()
plt.show()
```

Finalmente, podemos definir a __similaridade__ entre dois pontos a partir da distância entre eles:

    - Quanto mais distantes dois pontos, menos similares eles são;
    
    - Quanto menos distantes dois pontos, mais similares eles são;
    
Ou seja, a relação entre __similaridade__ e __distância__ é inversa!

Tendo um valor D de distância entre dois pontos $X$ e $Y$, podemos definir

### $Similaridade(X,Y) = \frac{1}{(1 + D)}$

<br><br>

### Obs.: 

Estamos, neste caso, definindo __uma medida de distância e de similaridade__ entre dois pontos. 

No entanto, várias outras medidas de similaridade/distância podem ser consideradas nos algoritmos!!

<br>

### Retomando para nosso problema objetivo..

Nosso próximo passo é implementar uma classe python que calcula a similaridade entre usuários e entre itens:

<br>

```python
class Similaridade():
    
    def __init__(self, dataframe):
        self.df = dataframe
        self.lista_usuarios = dataframe.index.tolist()
        self.lista_filmes = dataframe.columns.tolist()
        
    def similaridadeUsuarios(self, user, sorted_list = True):
        nota_user = self.df.loc[user, :].values 
        L = []
        for u in self.lista_usuarios:
            if u == user:
                continue
            else:
                nota_u = self.df.loc[u, :].values
                dist = euclid_dist(nota_user, nota_u)
                similaridade = 1 / (1 + dist)
                L.append([u,similaridade])
        if sorted_list:
            L.sort(key = lambda s: s[1], reverse = True)
            return L
        else:
            return L

    def similaridadeFilmes(self, filme, sorted_list = True): 
        nota_filme = self.df.T.loc[filme, :].values
        L = []
        for f in self.lista_filmes: 
            if f == filme:
                continue
            else:
                nota_f = self.df.T.loc[f, :].values
                dist = euclid_dist(nota_filme, nota_f)
                similaridade = 1 / (1 + dist)
                L.append([f,similaridade])
        if sorted_list:
            L.sort(key = lambda s: s[1], reverse = True)
            return L
        else:
            return L
        
    def similaridade_entre_dois_usuarios(self, user1, user2):
            return [x[1] for x in self.similaridadeUsuarios(user1) if x[0] == user2 ][0]

    def similaridade_entre_dois_filmes(self, filme1, filme2):
        return [x[1] for x in self.similaridadeFilmes(filme1) if x[0] == filme2 ][0]
```

## 3. Filtros colaborativos baseados em USUÁRIOS: implementação

Antes de implementarmos o __filtro colaborativo baseado em usuários,__ vamos implementar a classe InfoRec, que nos auxiliará com informações a respeito dos usuários e itens:

<br>

```python
class InfoRec():
    
    def __init__(self, dataframe):
        self.df = dataframe
        
    def getFilmes(self, user):
        return self.df.loc[user]

    def getUsers(self, filme):
        return self.df.T.loc[filme]

    def getPossibleMovies(self, user):
        return self.getFilmes(user)[np.isnan(self.getFilmes(user))].index.tolist()

    def getPossibleUsers(self, filme):
        return self.getUsers(filme)[np.isnan(self.getUsers(filme))].index.tolist()
```

### Fórmula:

<br>

Considere um usuário $u$ e um item (neste caso, um filme) $i$. Queremos calcular $p(u,i)$, ou seja, a __predição da nota do usuário $u$ no item $i$.__


A fórmula é dada por:


## $p(u,i) = \frac{\sum_{v; v \in N_{i}} S(u,v) \cdot R(v,i)}{\sum_{v; v \in N_{i}} S(u,v)}$, 

onde:

$S(u,v) $ é a similaridade entre os usuários $u$ e $v$.

$R(v,i)$ é a nota (rating) do usuário $v$ no item $i$.

{${v; v \in N_{i}}$} é o conjunto dos usuários que avaliaram o item $i$.

<br>

__Vamos aplicar esta fórmula para um usuário específico:__

<br>

```python
#fixamos um usuário específico

user = "User_1"
```

```python
#calculamos a similaridade de todos os outros usuários com este usuário específico

df1 = pd.DataFrame(sim.similaridadeUsuarios(user, sorted_list = False), 
                   columns = ['usuario', 'similaridade'])
df1
```

```python
#coletamos, para cada um dos outros usuários, as notas que eles deram para os filmes

df2 = sim.df.loc[df1.usuario, ].reset_index().rename(columns = {"index":"usuario"})
df2
```

```python
#agrupamos as duas tabelas acima numa única

df3 = df1.merge(df2, on = 'usuario', how = 'inner')
df3
```

```python
#extraindo quais são os filmes possíveis para indicar para este usuário específico

info.getPossibleMovies(user)
```

```python
#calculamos as predições para cada um dos filmes disponíveis


#dicionario para armazenar as predições de cada filme
dic_recs = {}

#loop nos filmes disponíveis
for f in info.getPossibleMovies(user): 
    
    #acessando a coluna de similaridades
    sims = df3.similaridade
    
    #acessando a coluna de notas do filme dentro do loop
    nota_f = df3[f]
    
    #fazendo as contas, de acordo com a fórmula
    total_sim_vs_notafilme = (sims * nota_f.fillna(0)).sum()
    total_sims = sims[~sims.index.isin(nota_f.loc[np.isnan(nota_f)].index.tolist())].sum()
    
    #salvando a predição final de cada filme no dicionário
    dic_recs[f] = total_sim_vs_notafilme / total_sims
    
dic_recs
```

<br>

__Colocando isso numa classe python:__
    
<br>

```python
class UserBasedColabFiltering():

    def __init__(self, dataframe):
        self.df = dataframe
        self.sim = Similaridade(dataframe = self.df)
        self.info = InfoRec(dataframe = self.df)
        
    def makeRec(self, user):
        df1 = pd.DataFrame(self.sim.similaridadeUsuarios(user, sorted_list = False), 
              columns = ['usuario', 'similaridade'])
        df2 = self.df.loc[df1.usuario, ].reset_index().rename(columns = {"index":"usuario"})
        df3 = df1.merge(df2, on = 'usuario', how = 'inner')
        dic_recs = {}
        for f in self.info.getPossibleMovies(user): 
            sims = df3.similaridade
            nota_f = df3[f]
            total_sim_vs_notafilme = (sims * nota_f.fillna(0)).sum()
            total_sims = sims[~sims.index.isin(nota_f.loc[np.isnan(nota_f)].index.tolist())].sum()
            dic_recs[f] = total_sim_vs_notafilme / total_sims
        return dic_recs
```

## 4. Filtros colaborativos baseados em ITENS: implementação

<br>

### Fórmula:

<br>


Considere um usuário $u$ e um item (neste caso, um filme) $i$. Queremos calcular $p(u,i)$, ou seja, a __predição da nota do usuário $u$ no item $i$.__


A fórmula é dada por:


## $p(u,i) = \frac{ \sum_{n; n \in I_u} S(i,n) \cdot R(u,n)  }   {  \sum_{n; n \in I_u} S(i,n) }$

$S(i,n) $ é a similaridade entre os itens $i$ e $n$.

$R(u,n)$ é a nota (rating) do usuário $u$ no item $n$.

{$n; n \in I_u$} é o conjunto de itens que o usuário $u$ avaliou.

<br><br>

No caso da filtragem por usuário, nossa técnica para cálculo das recomendações foi fixar um usuário e então calcular todas as recomendações de filmes para este usuário.

Neste caso em questão - filtragem por item - iremos fazer o contrário: iremos fixar um filme e analisar os usuários a respeito desse filme!

Na prática, o resultado final tem a mesma interpretação. Mas acabamos fazendo essa inversão durante a construção do algoritmo. Isso deve-se ao fato de que na filtragem baseada em usuários, as similaridades são calculadas entre os usuários; enquanto que na filtragem por itens a similaridade é a nível item.

<br>

```python
class ItemBasedColabFiltering():

    def __init__(self, dataframe):
        self.df = dataframe
        self.sim = Similaridade(dataframe = self.df)
        self.info = InfoRec(dataframe = self.df)
        
    def makeRec(self, filme):
        df1 = pd.DataFrame(self.sim.similaridadeFilmes(filme, sorted_list = False), 
                           columns = ['filme', 'similaridade'])
        df2 = self.df.T.loc[df1.filme, ].reset_index().rename(columns = {"index":"filme"})
        df3 = df1.merge(df2, on = 'filme', how = 'inner')
        dic_recs = {}
        for u in self.info.getPossibleUsers(filme):
            sims = df3.similaridade
            nota_u = df3[u]
            total_sim_vs_notafilme = (sims * nota_u.fillna(0)).sum()
            total_sims = sims[~sims.index.isin(nota_u.loc[np.isnan(nota_u)].index.tolist())].sum()
            dic_recs[u] = total_sim_vs_notafilme / total_sims
        return dic_recs
```

## 5. Filtros colaborativos baseados em modelos: Fatoração Matricial 


<br>

Novamente, estamos trabalhando no contexto de __filtros colaborativos,__ logo o objeto principal de nosso estudo é a __matriz de interação usuário/item.__

No caso do método de __Fatoração Matricial,__ o objetivo é:

    - Decompor a matriz de interação no produto de duas matrizes bem menores:
    
        - Matriz de fatoração dos usuários;
        
        - Matriz de fatoração dos itens;
        
        - A multiplicação entre essas duas matrizes deve gerar uma matriz resultante similar a matriz de interação original.
        
<br>
        
Neste caso, a matriz de interação é esparsa, como já discutimos. Após a fatoração, as duas matrizes resultantes __não serão esparsas!__

Desta forma, ao multiplicarmos as duas matrizes, teremos uma matriz final com o mesmo shape que a matriz de interação original, mas com todas as entradas preenchidas. 

### Conclusão: esses valores preenchidos (que até então estavam faltantes) serão justamente as recomendações!

<br>

__Matematicamente,__ a principal hipótese por trás da fatoração matricial é que existe um __espaço dimensional latente de baixa dimensão__ no qual podemos __representar__ tanto os __usuários__ quanto os __itens.__ Além disso, podemos __estimar a interação entre usuários e itens__ a partir deste produto matricial.

Na prática, iremos fixar um valor $K$, representando a quantidade de __variáveis latentes__, ou seja, a dimensão deste espaço acima citado.

Veja a figura abaixo:

(na figura, é assumido que temos $n$ usuários e $m$ itens. Além disso, a quantidade de variáveis latentes é $l$.)

<br><br>

### Matemática por trás da Fatoração Matricial (Opcional)

<br>

__(Essa seção é opcinal e o leitor conseguirá usar os métodos de fatoração matricial mesmo que não tenha o conhecimento aqui discutido)__

<br>

Nesta seção, o objetivo é discutir em mais detalhes matemáticos o funcionamento da fatoração matricial. Aqui, assumimos que o leitor tenha uma base matemática em cálculo e álgebra linear.

Especificamente, iremos discutir como utilizar o método de __gradient descent__ para obter a fatoração matricial.

<br>


Vamos denotar a matriz de interação usuários/itens por $R$. Considerando que temos N usuários e M itens, a dimensão de $R$ é [N,M]. 

Vamos ainda denotar por $K$ a quantidade de variáveis latentes.

O método de fatoração de matrizes consiste de escrever a matriz $R$ como produto de duas matrizes $P$ e $Q$.

Ou seja, estamos interessados em encontrar duas matrizes, $P$ e $Q$ tais que:

### $R  \sim  P \cdot Q := \hat R$

Desta forma, a dimensão de $P$ é [N, $K$] e a dimensão de $Q$ é [$K$, M].
Cada linha de $P$ corresponde a um usuário e as colunas são a __representação desses usuários__ no __espaço latente.__ Além disso, a transposta de $Q$ tem dimensão [M, $K$] e cada linha dessa matriz corresponde a __representação dos itens__ no espaço latente.


Uma vez com as matrizes $P$ e $Q$ em mãos, as predições são dadas pela multiplicação das mesmas, o que estamos denotando por $\hat R$. 

Em outras palavras, __a nota (rating) do usuário $i$ para o item $j$ é dada por:__

## $\hat r_{ij} = \sum_{k = 1, .., K} p_{ik} q_{kj}$,

correspondendo a i-ésima linha e a j-ésima coluna de $\hat R$.

__A pergunta que fica é: como encontrar as matrizes $P$ e $Q$?__

Vamos utilizar o método __gradient descent__ para isso!

Desta forma, na prática, começaremos com matrizes $P$ e $Q$ aleatórias e iremos iterativamente alterando as mesmas, com o objetivo de minimizar uma determinada função de custo.

Neste caso, podemos __definir o erro__ associado ao usuário $i$ e item $j$ sendo:

## $e_{ij}^2 = ( r_{ij} - \hat r_{ij} )^2 = ( r_{ij} - \sum_{k = 1, .., K} p_{ik} q_{kj} )^2$,

naturalmente, __este erro está definido apenas nos valores em que temos interações na matriz R!!__

<br>

Definindo a função de custo como:

## $J = \sum_{i} \sum_{j} e_{i,j}$

em outras palavras, a função de custo associada é essencialmente o __MSE!__

Além disso, veja que $J$ depende dos parâmetros $p_{ik}$ e $q_{kj}$ para $k \in $ {1,.., $K$}, $n \in $ {1,.., N} e $j \in $ {1,.., M}.

Desta forma, o gradiente de $J$, escrito pelas derivadas parciais, é:

## $\frac {\partial J}{\partial p_{ik}} = -2 e_{ij} q_{kj}$

## $\frac {\partial J}{\partial q_{kj}} = -2 e_{ij} p_{ik}$

Concluimos então que a __regra de atualização dos coeficientes__ é:

## $p_{ik}' = p_{ik} - \alpha \frac {\partial J}{\partial p_{ik}} = p_{ik} + 2 \alpha e_{ij} q_{kj}$

## $q_{kj}' = q_{kj} - \alpha \frac {\partial J}{\partial q_{kj}} = q_{kj} + 2 \alpha e_{ij} p_{ik}$

onde $\alpha$ é a taxa de aprendizagem (learning rate) do algoritmo.

<br>

__Implementando no python:__

```python
import time
```

```python
class MatrixFactorization():
    
    def __init__(self, dataframe, K, steps, alpha, beta):
        self.df = dataframe
        self.K = K
        self.steps = steps
        self.alpha = alpha
        self.beta = beta
        
    def fit(self):
        t0 = time.time()
        
        R = self.df.values
        N, M = R.shape
        
        #inicio aleatorio
        P = np.random.rand(N,self.K)
        Q = np.random.rand(self.K,M)
        
        lista_erro_step = []
        
        #loop
        for step in range(self.steps):
            
            mse_total_step = 0
            #varrendo todas as entradas da matriz R
            for i in range(len(R)):
                for j in range(len(R[i])):
                    #validando se o valor associado está preenchido
                    if R[i][j] > 0:

                        #calculando o erro:
                        eij = R[i][j] - np.dot(P[i,:],Q[:,j])
                        mse_total_step += (eij)**2
                        #alterando os valores
                        for k in range(self.K):
                            P[i][k] = P[i][k] + self.alpha * ( 2 * eij * Q[k][j] - self.beta * P[i][k])
                            Q[k][j] = Q[k][j] + self.alpha * ( 2 * eij * P[i][k] - self.beta * Q[k][j])
                            
            lista_erro_step.append(mse_total_step)
            
        self.P = P
        self.Q = Q
        self.lista_erro_step = lista_erro_step
        t1 = time.time()
        print("Fatoração concluída. Tempo aproximado:", int((t1-t0)/60)+1, 'minuto(s).')
        
    def predict(self):
        return self.P.dot(self.Q)
    
    def print_MSE_steps(self):
        plt.figure(figsize=[15,6])
        plt.title("Custo total por Step", fontsize = 16, fontweight = 'bold')
        plt.xlabel("Step", fontsize = 14, fontweight = 'bold')
        plt.ylabel("Erro", fontsize = 14, fontweight = 'bold')
        plt.plot(range(1, 1+self.steps), self.lista_erro_step, c = 'blue', lw = 2)
        plt.grid()
        plt.show()
```

## 6. Critérios de avaliação de Sistemas de Recomendação

<br>

Assim como os métodos de recomendação são vastos, existem também várias métricas atreladas a performance dos sistemas de recomendação.

Inclusive, links próprios para essa discussão serão disponibilizados no chat do curso.

Além disso, discutiremos uma metodologia para avaliação de filtros colaborativos nos exercícios deste módulo!

<br>

Na prática, o __melhor jeito de medir a performance de um sistema de recomendação é realmente avaliá-lo em produção,__ ou seja: colocamos o modelo em produção e __medimos os resultados em dados reais.__

No entanto, claramente, __existe um risco__ nessa abordagem: caso o modelo posto em produção gere recomendações ruins, isso terá um impacto direto na empresa!

Um jeito muito comum de tentar minimizar esse risco é realizando um __teste A/B.__

Essencialmente, em vez de simplesmente aplicarmos nosso modelo em produção para todos os usuários, faremos o seguinte:

    - Usaremos uma pequena parcela dos usuários (por exemplo, 1%, 2%, ..) para aplicar o modelo; (grupo A)
    
    - Usaremos outra parcela dos usuários e não aplicaremos o modelo; (grupo B)
    
   
Esses grupos (amostras) devem ser escolhidos de forma aleatória, para que sejam comparáveis entre si e que não tenhamos um viés de seleção dessas amostras.

Uma vez coletados os resultados nos grupos, podemos comparar a performance entre eles. Por exemplo, caso o modelo seja realmente melhor do que uma recomendação aleatória, devemos esperar que os resultados do grupo A (grupo com o modelo) sejam melhores do que os resultados do grupo B (grupo que teve recomendações aleatórias).

<br>
