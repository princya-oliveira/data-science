- Funco map

```python
# Para construir os modelos de ML transformamos todos os campos que so strings em campos numricos. Vejam o que
#podemos fazer, por exemplo, pra varivel sexo.

print(dados.sexo[:5])

dados.sexo = dados.sexo.map({"male":0,"female":1})

print(dados.sexo[:5])
```

- Groupby

```python
dados.head()
```

```python
#Podemos fazer agrupamentos aqui, assim como fizemos em SQL. Lembram da funo group by l?
dados.groupby(by = "sexo")
```

Esse groupby faz algo desse tipo aqui:

```python
dados[dados.sexo==1][:5]
```

```python
dados[dados.sexo==0][:5]
```

```python
# Uma expresso lambda permite escrever funes annimas, em tempo de cdigo, usando apenas uma linha de cdigo. 
# Onde x1,x2... so variveis que representam o argumento da funo e expr  qualquer expresso vlida em Python,
#envolvendo essas variveis -- lambda x,y: np.sum(x,y)

dados.groupby(by = "sexo").apply(lambda x: x.idade.mean() )
```

```python
dados.groupby(by = "sexo").apply(lambda x: x.Sobreviveu.value_counts() )
```

```python
# Poderiamos tambm aplicar uma funo nossa, customizada:

def f(x):
    return np.max(x) - np.min(x)

dados.groupby(by = "sexo").apply(lambda x: f(x.idade) )
```

```python
# Podemos usar o apply sem usar groupby por exemplo, apenas aplicar nas series diretamente:

dados[['Passagem', 'idade']].apply(lambda s: s.mean())
```

```python
#Outra funo muito usada tambm, alm da apply  a aggragate. Aqui usamos a agg para aplicar a funo min() em
#todas as colunas:

dados.groupby('Cabine').agg('min')
```

```python
# Podemos usar a agg para aplicar uma funo especfica para cada coluna de um dataframe:

def moda(s):
    ss = s.value_counts()
    return ss.idxmax()

dados.groupby('Cabine').agg({"Sobreviveu":'sum', "idade":['min', 'max'], "Embarcou": moda})
```
