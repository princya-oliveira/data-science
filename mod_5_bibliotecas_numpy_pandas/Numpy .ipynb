- Reshape

```python
# Alteramos o shape dos dados - remodelamos o array

a = np.arange(16)
print(a)
print('-------' * 7)
a = a.reshape((4, 4))
print(a)
```

```python
#Vocs vero nos prximos mdulos que as funes reshape(-1,1) e reshape (1,-1) so muito utilizadas:

b = np.array([0,0,3])
print("Antes do reshape:", b.shape)

# Sempre que colocamos o reshape (-1,1) ou (1,-1) a dimenso que j colocamos -1 se mantm e adicionamos uma nova 
#dimenso:

print("Aps reshape (-1,1): ", b.reshape(-1,1).shape)


b = b.reshape(1,-1)
print("Aps reshape (1,-1): ", b.shape)
```

- Slicing

Podemos "fatiar" os arrays, selecionando elementos especificos, assim como fizemos com as listas, da seguinte maneira:
**array[inicio:fim]**, sendo incio incluso e o fim no incluso. Caso no passemos o incio ele comear de 0 e caso no passemos o fim ele ir at o fim do array, ou seja, o ltimo elemento.


```python
#No caso unidimensional

print(arr)

#todos os elementos a partir do elemento cujo ndice  2.
print('----------------------------')
print(arr[2:])

#todos os elementos at o elemento cujo ndice  4 (Lembre-se que o elemento do ndice 4 no  incluso).
print('----------------------------')
print(arr[:4])

#todos os elementos entre a 2 e a 5 posio
print('----------------------------')
print(arr[2:5])
```

```python
#No caso bidimensional

print(arr_bidim)

#todas as linhas, coluna 1
print('----------------------------')
print(arr_bidim[:,1])

#todas as colunas, linha 0:
print('----------------------------')
print(arr_bidim[0,:])

#todas as linhas da coluna 1 at a 3:
print('----------------------------')
print(arr_bidim[:,1:3])
```

- Concatenando arrays

```python
# Funes utilizadas para  "colar" dois arrays

ar1 = np.arange(10,20,2)
ar2 = np.arange(20,30,2)

print(ar1)
print(ar2)
```

```python
# c_  uma funo para concatenar dois arrays -- muito utilizado na rea de DS para criao de bases de dados,
# pois geralmente pegamos variveis de diversas tabelas e nessa caso, podemos concatenar nada para trasnformar
#em uma nica tabela:

#mantm a ordem dos dados:

arr_c_ = np.c_[ar1,ar2]

print(arr_c_)

print(arr_c_.shape)
```

```python
#concatena dois arrays unidimensionais na horizontal:

arr_concatenate = np.concatenate([ar1,ar2])
print(arr_concatenate)
                                 
print(arr_concatenate.shape)
```

```python
arr1_bdim = np.array([[1,2],[6,4]])
arr2_bdim = np.array([[0,8],[3,3]])

print(arr1_bdim)
print('--------')
print(arr2_bdim)

# A vuno v-stack ela opera um vertical stack, ou seja, ela "cola" os arrays na vertical:

arr_v_stack = np.vstack((arr1_bdim, arr2_bdim))
print('--------')
print(arr_v_stack)
```

```python
# A vuno h-stack ela opera um horizontal stack, ou seja, ela "cola" os arrays na horizontal:

arr_h_stack = np.hstack((arr1_bdim, arr2_bdim))

print(arr_h_stack)
```
