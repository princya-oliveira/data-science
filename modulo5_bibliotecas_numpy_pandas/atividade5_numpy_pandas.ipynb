```python
#5.1: Crie uma função que receba um número inteiro positivo N e retorne um array 0, 1, ..., N.

n = 15

if n >= 0:
    arr_ativ = np.arange(n + 1)
    print(arr_ativ)
else:
    print("O número inserido não é um inteiro positivo.")
```

```python
#5.2: Crie uma função que receba um array [0,1, .., N] e retorne o mesmo array, só que com os números em posições aleatórias.

arr_ativ2 = np.array([0,1,2,3,4,5,6])
print("Array inicial:", arr_ativ2)

print("----------------------------")

np.random.shuffle(arr_ativ2)
print("Array posições aleatórias:", arr_ativ2)
```

```python
#5.3: Crie uma função que receba um número inteiro positivo N. Essa função deve retornar um array de tamanho N, 
#cujos valores são números inteiros entre 0 e N-1 e pode haver repetição desses números.

n = 6

if n >= 0:
    arr_ativ = np.random.choice(n - 1, size = n)
    print(arr_ativ)
else:
    print("O número inserido não é um inteiro positivo.")
```

__Exercício 6:__ Construa uma função divide_dados:
        
        - recebe um array X de dimensões quaisquer (n,k) e um parâmetro p entre 0 e 1.
        - divide o array X em duas partes disjuntas de forma aleatória. Ou seja, crie dois novos arrays, X1 e X2,
        onde a dimensão de linhas de X1 somada com a dimensão de linhas de X2 deve ser igual a dimensão de linhas de X.
        - interpretando p como percentual, o tamanho de X1 deve ser p*100% do número de linhas de X; enquanto X2 deve ser o resto.
        - retorne esses dois arrays.

```python
x = np.random.rand(10, 2) #array x -> 10 linhas e 2 colunas
p = 70 #percentual entre 0 e 100

size_x = len(x) #tamanho de x
size_x1 = p*(size_x/100) #cálculo tamanho de x1
size_x2 = size_x - size_x1 #cálculo tamanho de x2

#dividindo o array em 2
x1_idx = np.random.randint(x.shape[0], size=int(size_x1)) 
x2_idx = np.random.randint(x.shape[0], size=int(size_x2))
x1, x2 = x[x1_idx,:], x[x2_idx,:]

print("x1:", x1)
print("**************")
print("x2:", x2)
```
