#### Ainda falando de slicing, podemos citar duas funções muito utilizadas: LOC e ILOC

**loc**: Usamos para filtrar informação no DF e funciona da seguinte maneira: df.loc[linhas,colunas]. Selecionamos as colunas pelo nome.
   
**iloc**: Funciona de maneira similar:  df.iloc[linhas, colunas], porém  o iloc seleciona linhas e colunas por números (indices).

```python
#todas as linhas e apenas duas colunas 
dados.loc[:, ["Sobreviveu","idade"]]
```

- Funcão map

```python
# Para construir os modelos de ML transformamos todos os campos que são strings em campos numéricos. Vejam o que
#podemos fazer, por exemplo, pra variável sexo.

print(dados.sexo[:5])

dados.sexo = dados.sexo.map({"male":0,"female":1})

print(dados.sexo[:5])
```

- Groupby

```python
dados.head()
```

```python
#Podemos fazer agrupamentos aqui, assim como fizemos em SQL. Lembram da função group by lá?
dados.groupby(by = "sexo")
```

Esse groupby faz algo desse tipo aqui:

```python
dados[dados.sexo==1][:5]
```

```python
dados[dados.sexo==0][:5]
```

```python
# Uma expressão lambda permite escrever funções anônimas, em tempo de código, usando apenas uma linha de código. 
# Onde x1,x2... são variáveis que representam o argumento da função e expr é qualquer expressão válida em Python,
#envolvendo essas variáveis -- lambda x,y: np.sum(x,y)

dados.groupby(by = "sexo").apply(lambda x: x.idade.mean() )
```

```python
dados.groupby(by = "sexo").apply(lambda x: x.Sobreviveu.value_counts() )
```

```python
# Poderiamos também aplicar uma função nossa, customizada:

def f(x):
    return np.max(x) - np.min(x)

dados.groupby(by = "sexo").apply(lambda x: f(x.idade) )
```

```python
# Podemos usar o apply sem usar groupby por exemplo, apenas aplicar nas series diretamente:

dados[['Passagem', 'idade']].apply(lambda s: s.mean())
```

```python
#Outra função muito usada também, além da apply é a aggragate. Aqui usamos a agg para aplicar a função min() em
#todas as colunas:

dados.groupby('Cabine').agg('min')
```

```python
# Podemos usar a agg para aplicar uma função específica para cada coluna de um dataframe:

def moda(s):
    ss = s.value_counts()
    return ss.idxmax()

dados.groupby('Cabine').agg({"Sobreviveu":'sum', "idade":['min', 'max'], "Embarcou": moda})
```
